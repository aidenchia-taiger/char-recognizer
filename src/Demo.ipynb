```python
X = []
Y = []
import os
import cv2
for root, _, files in os.walk('../imgs/validation/'):
    for file in files:
        if '.png' not in file:
            continue
        X.append(cv2.imread(os.path.join(root, file), cv2.IMREAD_GRAYSCALE))
        Y.append(root.split('/')[-1])
```

```python
from keras.preprocessing.image import ImageDataGenerator
valid_datagen = ImageDataGenerator(rescale=1./255)
valid_generator = valid_datagen.flow_from_directory('../imgs/validation', batch_size=1, color_mode='grayscale',
                                                    target_size =(32,32), shuffle=False)
```

```python
mapping = valid_generator.class_indices
inv_mapping = {v:k for k,v in mapping.items()}
#inv_mapping
inv_mapping
```

```python
i = 0
while i < len(valid_generator):
    x,y = valid_generator.next()
    img = x * 255
    img = img.astype(int)
    img = np.squeeze(img, axis=0)
    img = np.squeeze(img, axis=-1)
    pred = inv_mapping[np.argmax(model.predict(x))]
    gt = inv_mapping[np.argmax(y)]
    print('Ground Truth: {}, Prediction: {}'.format(gt, pred))
    if gt != pred:
        cv2.imwrite('../analysis/GT{}_P{}_{}.png'.format(gt,pred,i), img)
    i+=1
```
