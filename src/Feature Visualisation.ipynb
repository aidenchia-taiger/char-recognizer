```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
from Distinguisher import *
```

```python
%load_ext autoreload
%autoreload 2
```

```python
path = '../dist_imgs/train/'
```

```python
dist = Distinguisher()
```

```python
[X, y], paths = loadData(dist, path, getPaths=True)
```

```python
features = ['alignment', 'uniformity', 'otsu', 'variance', 'density', 'mean', 'peaks', 'straightness']
```

```python
df = pd.DataFrame(X, None, columns=features)
df.head()
```

```python
df['gt'] = y
df.head(10)
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
ax = seaborn.scatterplot(x="alignment", y="mean", hue= "gt",data=df)
plt.axvline(x=30, color='r')
plt.axhline(y=150, color='r')
plt.axhline(y=210, color='r')
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
ax = seaborn.scatterplot(x="uniformity", y="otsu", hue= "gt",data=df)
plt.axvline(x=15, color='r')
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
ax = seaborn.scatterplot(x="density", y="variance", hue= "gt",data=df)
plt.axvline(x=0.55, color='r')
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
ax = seaborn.scatterplot(x="peaks", y="straightness", hue= "gt",data=df)
plt.axhline(y=5, color='r')
plt.axhline(y=90, color='r')
```
